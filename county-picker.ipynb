```python
FIPS['fips']=FIPS['fips_code'].astype(str)
FIPS.head()
```

```python
FIPS['selected']=0
FIPS.loc[FIPS['locality']=='Accomack', 'selected']=1
FIPS.head()
```

```python
print(FIPS['fips'].dtype)
print(FIPS['selected'].dtype)
```

```python
import plotly.figure_factory as ff

values = FIPS['selected']
fips = FIPS['fips']

fig = ff.create_choropleth(fips=fips, 
                           values=values, 
                           colorscale=['Gray', 'Yellow'],
                           scope=['VA'], 
                           county_outline={'color': 'rgb(255,255,255)', 'width': 0.5})
iplot(fig)
```
